```python
#Handle Missing Values
df.fillna(0, inplace=True)
```

```python
#Remove Duplicates
df.drop_duplicates(inplace=True)
```

```python
# calculate the mean value for "children" column
avg_children = df["children"].astype("float").mean(axis = 0)
print("Average of children:", avg_children)

# replace NaN by mean value in "children" column
df["children"].replace(np.nan, avg_children, inplace = True)
```

```python
# calculate the mean value for "agent" column
avg_agent = df["agent"].astype("float").mean(axis = 0)
print("Average of agent:", avg_agent)

# replace NaN by mean value in "agent" column
df["agent"].replace(np.nan, avg_agent, inplace = True)
```

```python
# calculate the mean value for 'adr' column
avg_adr = df["adr"].astype("float").mean(axis = 0)
print("Average of adr:", avg_adr)

# replace NaN by mean value
df["adr"].replace(np.nan, avg_adr, inplace = True)
```

```python
#For categorical columns like country, you should use mode (not mean)
# calculate most frequent value (mode)
mode_country = df["country"].mode()[0]
print("Most frequent country:", mode_country)

# replace NaN with mode
df["country"].replace(np.nan, mode_country, inplace = True)
```

```python
# simply drop whole row with NaN in "children" column
df.dropna(subset=["children"], axis=0, inplace=True)
```

```python
# reset index after dropping rows
df.reset_index(drop=True, inplace=True)
```
